```python
# Netflix EDA - Updated for actual dataset columns

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Administrator\Desktop\Exploratory Data Analysis\Netflix Engagement\dataset\Netflix_Engagement_Report_2023_Jan-Jun_cleaned.csv')

display(df.head())

print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

print("\nDuplicate rows:", df.duplicated().sum())

print("\nDescriptive Statistics:\n")
display(df.describe(include='all'))

top_titles = df.sort_values('Hours Viewed', ascending=False).head(10)
sns.barplot(data=top_titles, x='Hours Viewed', y='Title', palette='viridis')
plt.title('Top 10 Netflix Titles by Hours Viewed (Jan-Jun 2023)')
plt.xlabel('Hours Viewed')
plt.ylabel('Title')
plt.tight_layout()
plt.show()

sns.histplot(df['Hours Viewed'], bins=30, kde=True)
plt.title('Distribution of Hours Viewed')
plt.xlabel('Hours Viewed')
plt.ylabel('Frequency')
plt.show()

df['Month'] = df['Release Date'].dt.to_period('M')
monthly_trend = df.groupby('Month')['Hours Viewed'].sum()

monthly_trend.plot(marker='o')
plt.title('Total Hours Viewed by Release Month')
plt.xlabel('Month')
plt.ylabel('Total Hours Viewed')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
